```python
import numpy as np
import pandas as pd

import matplotlib as mpl
import matplotlib.pyplot as plt

from matplotlib.animation import FuncAnimation

from sklearn.datasets import load_boston
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler

from IPython.display import HTML
```

```python
#Load the Dataset
boston=load_boston()

#Print Description of Dataset
print(boston.DESCR)
```

```python
#Put the data in pandas dataframe
features=pd.DataFrame(boston.data, columns=boston.feature_names)
features
```

```python
target=pd.DataFrame(boston.target, columns=['target'])
target
```

```python
#Concatenate features and target into a single frame. 
#Axis =1 putting the columns beside each other and not on top of each other i.e column-wise
df=pd.concat([features,target],axis=1)
df
```

```python
#use round(decimals=2) to set precision to 2 decimal places
df.describe().round(decimals=2)
```

```python
#Calculate correlation between every column of the dataset
corr = df.corr('pearson')

#Take absolute values of correlations
corrs = [abs(corr[attr]['target']) for attr in list(features)]

#Make list of pairs [(corr, feature)]
l = list(zip(corrs, list(features)))

# Sort the list of pairs in reverse/descending order
# with the correlation value as the key for sorting
l.sort(key = lambda x : x[0], reverse = True)

# Unzip pairs to two lists
# zip(*l) takes a list that looks like [[a,b,c], [d,e,f], [g,h,i]]
# and returns [[a,d,g], [b,e,h], [c,f,i]]
corrs, labels = list(zip((*l)))

#Plot correlations with respect to the target variable as a bar graph
index = np.arange(len(labels))
plt.figure(figsize=(15, 5))
plt.bar(index, corrs, width=0.5)
plt.xlabel('Attributes')
plt.ylabel('Corelation with the Target Variable')
plt.xticks(index, labels)
plt.show()
 
```

```python
#Normalize the Data between 0 & 1

X = df['LSTAT'].values
Y = df['target'].values

x_scaler = MinMaxScaler()
X = x_scaler.fit_transform(X.reshape(-1,1))
X = X[:, -1]
y_scaler = MinMaxScaler()
Y = y_scaler.fit_transform(Y.reshape(-1,1))
Y = Y[:, -1]
```

```python
# Define Error Function (Mean squared error)
def error(m, x, c, t):
    N = x.size
    e = sum(((m * x + c) - t) ** 2)
    return e * 1/(2 * N)
```

```python
#0.2 indicates 20% of the total data is randomply sampled as testing data
xtrain, xtest, ytrain, ytest = train_test_split(X, Y, test_size=0.2)
```

```python
#update function
def update(m, x, c, t, learning_rate):
    grad_m = sum(2 * ((m * x + c) - t) * x)
    grad_c = sum(2 * ((m * x + c) - t))
    m = m - grad_m * learning_rate
    c = c - grad_c * learning_rate
    return m, c
```

```python
#gradient_descent function

def gradient_descent(init_m, init_c, x, t, learning_rate, iterations, error_threshold):
    m = init_m
    c = init_c
    error_values =  list()
    mc_values = list()
    for i in range(iterations):
        e = error(m, x, c, t)
        if e < error_threshold:
            print('Error less than threshold. Stopping Gradient Descent.')
            break
        error_values.append(e)
        m, c = update(m, x, c, t, learning_rate)
        mc_values.append((m, c))
    return m, c, error_values, mc_values
```

```python
%%time
init_m = 0.5
init_c = 0
learning_rate = 0.001
iterations = 30000
error_threshold = 0.001

m, c, error_values, mc_values = gradient_descent(init_m, init_c, xtrain, ytrain, learning_rate, iterations, error_threshold)
```

```python
#As the number of iterations increases, changes in the line are more noticeable
# in order to reduce processing time, step size of the values is increased

mc_values_anim = mc_values[0:500:2]
```

```python
#Animation Plot of the Linear Regression Line

fig, ax = plt.subplots()
ln, = plt.plot([], [], 'ro-', animated=True)

def init():
    plt.scatter(xtest, ytest, color='g')
    ax.set_xlim(0, 1.0)
    ax.set_ylim(0, 1.0)
    return ln,

def update_frame(frame):
    m, c = mc_values_anim[frame]
    x1, y1 = -0.5, m * -0.5 + c
    x2, y2 = 1.5, m * 1.5 + c
    ln.set_data([x1, x2], [y1, y2])
    return ln,

anim = FuncAnimation(fig, update_frame, frames=range(len(mc_values_anim)), init_func=init, blit=True)

HTML(anim.to_html5_video())

    
```

```python
#Plotting the Regression-line over the training data
plt.scatter(xtrain, ytrain, color='b')
plt.plot(xtrain, (m * xtrain + c), color='r')

#Plotting Error in Iterations
plt.plot(np.arange(len(error_values)), error_values)
plt.ylabel('Error')
plt.xlabel('Iterations')
```
